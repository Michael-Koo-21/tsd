```python
import pandas as pd
from be_great import GReaT

# Create simple test data
df = pd.DataFrame({
    'age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'income': [30000, 45000, 55000, 60000, 70000, 75000, 80000, 85000, 70000, 50000],
    'education': ['HS', 'BA', 'BA', 'MA', 'MA', 'PhD', 'BA', 'HS', 'BA', 'HS']
})

print("Training GReaT model (this takes 1-2 minutes)...")
model = GReaT(
    llm='distilgpt2',  # Small model for testing
    epochs=5,          # Few epochs for quick test
    batch_size=2
)
model.fit(df)

print("\nGenerating synthetic data...")
synthetic = model.sample(n_samples=5)
print(synthetic)
print("\n✓ GReaT test successful!")
```

```python
import pandas as pd
from be_great import GReaT

# Load your data
df_train = pd.read_csv('great_train_data.csv')
print(f"Loaded {len(df_train)} training records")
print(f"Columns: {list(df_train.columns)}")

# Train GReaT (takes ~10-20 minutes on T4)
print("\nTraining GReaT model...")
model = GReaT(
    llm='distilgpt2',
    epochs=100,        # More epochs for real data
    batch_size=32,
    save_steps=5000
)
model.fit(df_train)
print("✓ Training complete!")

# Generate synthetic data (same size as training)
n_samples = len(df_train)
print(f"\nGenerating {n_samples} synthetic records...")
df_synthetic = model.sample(n_samples=n_samples)

# Save results
df_synthetic.to_csv('great_synthetic_data.csv', index=False)
print(f"✓ Saved great_synthetic_data.csv")

# Quick validation
print(f"\nValidation:")
print(f"  Shape: {df_synthetic.shape}")
print(f"  Columns match: {set(df_train.columns) == set(df_synthetic.columns)}")
print(df_synthetic.head())
```
